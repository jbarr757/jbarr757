```python
# Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configurtion of the plots we will crete


# Read the data

df = pd.read_csv('coin_eth.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
#Data types were working with

df.dtypes
```

```python
bt = df.corr()
```

```python
sns.heatmap(bt)
```

```python
# Higher Marketcap is related to higher close
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix - Numeric Features')

plt.xlabel('Ethereum Prices')

plt.ylabel('Ethereum Prices')

plt.show()
```

```python
# Scatter plot with Marketcap V Close


plt.scatter(x = df['Marketcap'], y = df['Close'])

plt.title('Marketcap vs Close')

plt.xlabel('Marketcap')

plt.ylabel('Close')

plt.show()


sns.regplot(x='Marketcap', y= 'Close', data= df, scatter_kws={"color": 'black'}, line_kws={"color": 'red'})
```
